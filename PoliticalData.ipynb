```python
import pyreadstat
import pandas as pd
import io

# Assuming your file is named 'my_data.sav'
file_name = 'NPORS_2024_for_public_release (1).sav'

# Write the uploaded file content to a file
with open(file_name, 'wb') as f:
  f.write(uploaded[file_name])

# Read the .sav file. This function returns the data as a pandas DataFrame
# and a metadata object.
df, meta = pyreadstat.read_sav(file_name)

# Display the first few rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(df.head())

# Access and print the variable labels from the metadata
print("\nVariable Labels:")
variable_labels = meta.column_labels

# Check if variable_labels is a dictionary before iterating
if isinstance(variable_labels, dict):
    for var_name, label in variable_labels.items():
        print(f"{var_name}: {label}")
elif isinstance(variable_labels, list):
    print("No variable labels found in the metadata.")
else:
    print(f"Unexpected type for variable_labels: {type(variable_labels)}")
```

```python
import fitz # PyMuPDF

# Assuming the uploaded PDF is named 'NPORS24 paper.pdf'
pdf_file_name = 'NPORS24 paper.pdf'
extracted_text = ""

try:
    with fitz.open(pdf_file_name) as doc:
        for page_num in range(doc.page_count):
            page = doc.load_page(page_num)
            extracted_text += page.get_text()
except FileNotFoundError:
    extracted_text = f"Error: The file '{pdf_file_name}' was not found."
except Exception as e:
    extracted_text = f"An error occurred: {e}"

print(extracted_text)
```
